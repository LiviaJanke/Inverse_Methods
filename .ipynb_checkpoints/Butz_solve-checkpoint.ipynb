Define forward model for (n-1)th order polynomial
\begin{equation}
p(t)=\sum_0^{n-1}(a_j t^j)
\end{equation}
which implies the forward model
\begin{equation}
K_{i,j}=t_i^j
\end{equation}
with the state vector
\begin{equation}
x_j = a_j.
\end{equation}



```python
# Define forward model for a (n-1)th order polynomial
def forward_poly(t,n):
    """
    # Forward model for (n-1)th order polynomial
    #
    # arguments: 
    #            t: independent real number
    #            n: order of polynomial, integer
    #
    # returns: 
    #            Ki: forward model for independent variable t,
    #                it is typically a row of the forward model matrix, np.array 
    """
    # check whether input is in range
    if not isinstance(n, int):
       raise SystemExit("ERROR! forward_poly: polynomial degree needs to be integer.")
    
    Ki = np.array([t**j for j in range(n)])
    
    return Ki
```

Calculate forward model:

```python
n = 5 # order of polynomial (n-1)
m = len(year) # length of the observation vector

y = co2 # observations: CO2 record from MLO
K = np.zeros([m,n]) # forward model (mxn)
?forward_poly
for i in range(m):
    K[i,:] = forward_poly(year[i],n)

# Rank of the forward model  
print('rank of the forward model =',np.linalg.matrix_rank(K))
```

Determine least-squares solution and calculate simulated observations:

```python
# Least-squares solution
xlsq, res, rank, svd = np.linalg.lstsq(K,y,rcond=-1)
# Simulated observations based on least-squares solution
ylsq = K.dot(xlsq)
```

Plot fit together with MLO observations and the residuals (meas-sim):

```python
# Plot MLO timeseries, fit
fig = plt.figure(figsize=(width, height))
plt.plot(year,co2,'k-',label='MLO CO$_2$')
plt.plot(year,ylsq,'r-', label='Fit (polynomial degree: %d)'%(n-1))
plt.ylabel('CO$_2$ / ppm', fontsize=fs)
plt.xlabel('Year', fontsize=fs)
plt.legend(fontsize=fs);

# Plot residuals: observations minus simulated
fig = plt.figure(figsize=(width, height))
plt.plot(year,y-ylsq,'k-',label='Residuals')
plt.ylabel('meas - sim (CO$_2$) / ppm', fontsize=fs)
plt.xlabel('Year', fontsize=fs)
plt.legend(fontsize=fs);
```

(Optional - If you want to do better.) Define forward model for n-th order Fourier series
\begin{equation}
f(t)=\sum_{j=1}^{n}(a_j cos(2\pi\,j\,t)+b_j sin((2\pi\,j\,t))
\end{equation}
which implies the forward model
\begin{equation}
K_{i,j}=cos(2\pi\,j\,t_i)
\end{equation}
for the a coefficients and
\begin{equation}
K_{i,n+j}=sin(2\pi\,j\,t_i)
\end{equation}
for the b coefficients which we put into the n+j columns of the forward model. The corresponding state vector is
\begin{eqnarray}
x_j &=& a_j\\
x_{n+j} &=& b_j.
\end{eqnarray}
Why is the above formulation of the problem ``easier'' than the conceptually equivalent 
\begin{equation}
f(t)=\sum_{j=1}^{n} a_j\,cos(2\pi\,j\,t + \varphi_j)?
\end{equation}
(Linear vs non-linear forward model.)


```python
# Define forward model for n-th order Fourier series
def forward_fourier(t,n):
    """
    # Forward model for n-th order Fourier series 
    # Sum_j=1^n( a_j*cos(2pi*j*t)) + b_j*sin(2pi*j*t) )
    #
    # arguments: 
    #            t: independent real number 
    #            n: order of Fourier series, integer
    #
    # returns: 
    #            Ki: forward model for independent variable t,
    #                it is typically a row of the forward model matrix, np.array (dim: 2n)
    """
    # check whether input is in range
    if not isinstance(n, int):
       raise SystemExit("ERROR! forward_fourier: Fourier series degree needs to be integer.")
    
    Ki = np.zeros(2*n) # 2n coefficients: n*a_j+n*b_j
    for j in range(n):
        Ki[j]= np.cos(2.*np.pi*float(j+1)*t)
        Ki[n+j] = np.sin(2.*np.pi*float(j+1)*t)
    return Ki
```

Calculate forward model:

```python
npoly = 5 # order of polynomial (n-1)
nfour = 3 # order of Fourier series

n = npoly + 2*nfour # length of state vector
m = len(year)       # length of observation vector

y = co2             # observations: CO2 record from MLO
K = np.zeros([m,n]) # forward model (mxn)
for i in range(m):
    K[i,0:npoly] = forward_poly(year[i],npoly)
    K[i,npoly:n] = forward_fourier(year[i],nfour)
```

Calculate least-squares solution and watch how the solution and the rank of the forward model changes when modifying the condition number cut-off of the built-in least-squares function. We will come back to what the condition number is later in the lecture series.

```python
# Least-squares solution
xlsq1, res, rank, svd = np.linalg.lstsq(K,y,rcond=None)
print('Attempt1: rank =', rank)
xlsq2, res, rank, svd = np.linalg.lstsq(K,y,rcond=1E-13)
print('Attempt2: rank =', rank)
xlsq3, res, rank, svd = np.linalg.lstsq(K,y,rcond=1E-14)
print('Attempt3: rank =', rank)
```

Plot the solutions and the residuals:

```python
fig = plt.figure(figsize=(width,height))
plt.plot(year,y,'k-', label='CO$_2$ MLO')
plt.plot(year,K.dot(xlsq1),'r-', label='rcond = standard')
plt.plot(year,K.dot(xlsq2),'g--', label='rcond = 1E-13')
plt.plot(year,K.dot(xlsq3),'b-', label='rcond = 1E-14')
plt.ylabel('CO$_2$ / ppm', fontsize=fs)
plt.xlabel('Year', fontsize=fs)
plt.xlim(1980,2000)
plt.legend(fontsize=fs);

fig = plt.figure(figsize=(width,height))
plt.plot(year,y-K.dot(xlsq1),'r-', label='rcond = standard')
plt.plot(year,y-K.dot(xlsq2),'g--', label='rcond = 1E-13')
plt.plot(year,y-K.dot(xlsq3),'b-', label='rcond = 1E-14')
plt.ylabel('meas - sim (CO$_2$) / ppm', fontsize=fs)
plt.xlabel('Year', fontsize=fs)
plt.legend(fontsize=fs);
```
